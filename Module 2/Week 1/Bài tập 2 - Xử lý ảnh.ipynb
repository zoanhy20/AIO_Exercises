```python
import matplotlib.image as mpimg
import numpy as np

img = mpimg.imread('dog.jpeg')

def color2grayscale(vector):
    return np.max(vector) * 0.5 + np.min(vector) * 0.5

gray_scale_01 = np.apply_along_axis(color2grayscale, 2, img)

plt.imshow(gray_scale_01, cmap='gray')
plt.axis('off')  # Optional: Turn off the axis
plt.show()
```

```python
import matplotlib.image as mpimg
import numpy as np

# Load the image
img = mpimg.imread('dog.jpeg')

# Convert to grayscale using the Lightness method
def color2grayscale(vector):
    return np.sum(vector) / 3

gray_scale_02 = np.apply_along_axis(color2grayscale, 2, img)

plt.imshow(gray_scale_02, cmap='gray')
plt.axis('off')  # Optional: Turn off the axis
plt.show()


# Print the value at position [0, 0]
print(gray_scale_02[0, 0])
```

```python
import matplotlib.image as mpimg
import numpy as np

# Load the image
img = mpimg.imread('dog.jpeg')

def color2grayscale(vector):
    return vector[0] * 0.21 + vector[1] * 0.72 + vector[2] * 0.07

gray_scale_04 = np.apply_along_axis(color2grayscale, 2, img)

plt.imshow(gray_scale_04, cmap='gray')
plt.axis('off')  # Optional: Turn off the axis
plt.show()


# Print the value at position [0, 0]
print(gray_scale_04[0, 0])
```
